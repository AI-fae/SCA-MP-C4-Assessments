```python

df.drop(["systemloanid", "customerid"], axis = 1, inplace = True )
```

```python
df.isnull().sum()
```

```python
#replacing Null values in numerical columns 
median1 = df["longitude_gps"].median()
median2 = df["latitude_gps"].median()
median3 = df["loannumber"].median()
median4 = df["loanamount"].median()
median5 = df["totaldue"].median()
median6 = df["termdays"].median()

df["termdays"].replace(np.nan, median6, inplace=True)
df["totaldue"].replace(np.nan, median5, inplace=True)
df["loanamount"].replace(np.nan, median4, inplace=True)
df["loannumber"].replace(np.nan, median3, inplace=True)
df["latitude_gps"].replace(np.nan, median2, inplace=True)
df["longitude_gps"].replace(np.nan, median1, inplace=True)
```

```python
mode1 = df["bank_account_type"].mode().values[0]
mode2 = df["bank_name_clients"].mode().values[0]
mode3 = df["bank_branch_clients"].mode().values[0]
mode4 = df["employment_status_clients"].mode().values[0]
mode5 = df["good_bad_flag"].mode().values[0]
#mode6 = df["level_of_education_clients"].mode().values[0] 

df["bank_account_type"].replace(np.nan, mode1, inplace=True)
df["bank_name_clients"].replace(np.nan, mode2, inplace=True)
df["bank_branch_clients"].replace(np.nan, mode3, inplace=True)
df["employment_status_clients"].replace(np.nan, mode4, inplace=True)
df["good_bad_flag"].replace(np.nan, mode5, inplace=True)

```

```python
df["level_of_education_clients"].replace("False", "Graduate", inplace=True)
#df["level_of_education_clients"].head(30)
df.info()
```

```python
mode6 = df["birthdate"].mode().values[0]
mode7 = df["approveddate"].mode().values[0]
mode8 = df["creationdate"].mode().values[0]
mode9 = df["closeddate"].mode().values[0]
mode10 = df["firstduedate"].mode().values[0]
mode11 = df["firstrepaiddate"].mode().values[0]
mode12 = df["creationdate"].mode().values[0]

df["birthdate"].replace(np.nan, mode6, inplace=True)
df["approveddate"].replace(np.nan, mode7, inplace=True)
df["creationdate"].replace(np.nan, mode8, inplace=True)
df["closeddate"].replace(np.nan, mode9, inplace=True)
df["firstduedate"].replace(np.nan, mode10, inplace=True)
df["firstrepaiddate"].replace(np.nan, mode11, inplace=True)
df["creationdate"].replace(np.nan, mode12, inplace=True)
```
